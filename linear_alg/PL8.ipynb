```python
# Пример 1 Построить эллиптический параболоид x^2 + 2y^2 = z
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from sympy import *

x = np.linspace(-1,1,200)
y = np.linspace(-1,1,200)

[x,y] = np.meshgrid(x,y)

z = lambda x,y: x**2 + 2*y**2 # Уравнение параболоида
Z = z(x,y)

fig = plt.figure(figsize =(10, 10))
ax = plt.axes(projection ='3d')
my_cmap = plt.get_cmap('magma') # Создание цветовой карты

# Создание графиков
surf = ax.plot_surface(x,y,Z,rstride = 8,cstride = 8,alpha = 0.8,cmap = my_cmap)

cset = ax.contourf(x,y,Z,zdir ='z',offset = np.min(Z),cmap=my_cmap)
cset = ax.contourf(x,y,Z,zdir ='x',offset =-2,cmap = my_cmap)
cset = ax.contourf(x,y,Z,zdir ='y',offset = 2,cmap = my_cmap)

fig.colorbar(surf,ax = ax,shrink = 0.5,aspect = 5)
ax.set_xlabel('X-axis')
ax.set_xlim(-2, 2)

ax.set_ylabel('Y-axis')
ax.set_ylim(-2, 2)

ax.set_zlabel('Z-axis')
ax.set_zlim(np.min(Z), np.max(Z))

ax.set_title('График эллиптического параболоида $x^2+2y^2=z$ с проекциями')
#ax.view_init(60, 30)
plt.show()

```

Упражнение 8.1. Построить поверхность x^2-y^2=z. Написать уравнения проекций данной поверхности на координатные плоскости.


```python
x = np.linspace(-1,1,200)
y = np.linspace(-1,1,200)

[x,y] = np.meshgrid(x,y)

z = lambda x,y: x**2 - y**2 # Уравнение параболоида
Z = z(x,y)

fig = plt.figure(figsize =(10, 10))
ax = plt.axes(projection ='3d')
my_cmap = plt.get_cmap('magma') # Создание цветовой карты

# Создание графиков
surf = ax.plot_surface(x,y,Z,rstride = 8,cstride = 8,alpha = 0.8,cmap = my_cmap)

cset = ax.contourf(x,y,Z,zdir ='z',offset = np.min(Z),cmap=my_cmap)
cset = ax.contourf(x,y,Z,zdir ='x',offset =-2,cmap = my_cmap)
cset = ax.contourf(x,y,Z,zdir ='y',offset = 2,cmap = my_cmap)

fig.colorbar(surf,ax = ax,shrink = 0.5,aspect = 5)
ax.set_xlabel('X-axis')
ax.set_xlim(-2, 2)

ax.set_ylabel('Y-axis')
ax.set_ylim(-2, 2)

ax.set_zlabel('Z-axis')
ax.set_zlim(np.min(Z), np.max(Z))

ax.set_title('График эллиптического параболоида $x^2-y^2=z$ с проекциями')
ax.view_init(40, 30) 
plt.show()
```

Упражнение 8.2. Построить сферу, используя ее параметрические уравнения:<br>


```python
fig = plt.figure(figsize =(10, 10))
ax = plt.axes(projection ='3d')

#Коэффициенты уравнения
coefs = (2, 2, 2)

rx, ry, rz = np.sqrt(coefs)

u = np.linspace(0, 2*np.pi, 200) # Угол на плоскости
v = np.linspace(-np.pi/2, np.pi/2, 200) # Угол в пространстве

#Уравнение эллипсоида
x = rx * np.outer(np.cos(u), np.cos(v))
y = ry * np.outer(np.sin(u), np.cos(v))
z = rz * np.outer(np.ones_like(u), np.sin(v))

my_cmap = plt.get_cmap('rainbow')

surf = ax.plot_surface(x,y,z,rstride = 10,cstride = 10,alpha = 0.6,cmap = my_cmap)

cset = ax.contourf(x,y,z,zdir ='z',offset = -2,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='x',offset =-4,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='y',offset = 2,cmap = my_cmap)

ax.set_xlabel('X-axis'); ax.set_xlim(-4, 4)
ax.set_ylabel('Y-axis'); ax.set_ylim(-2, 2)
ax.set_zlabel('Z-axis'); ax.set_zlim(-2, 2)

ax.set_title('График эллипсоида  $x^2/2+y^2/2+z^2/2=1$ с проекциями')

ax.view_init(15, -40)

plt.show()
```

Упражнение 8.3. Построить однополостный гиперболоид, используя его параметрические уравнения:<br>


```python
fig = plt.figure(figsize =(10, 10))
ax = plt.axes(projection ='3d')

#Коэффициенты уравнения
coefs = (0.5, 0.5, 0.5)

rx, ry, rz = np.sqrt(coefs)

u = np.linspace(0, 2*np.pi, 200) # Угол на плоскости
v = np.linspace(-np.pi/2, np.pi/2, 200) # Угол в пространстве

#Уравнение эллипсоида
x = rx * np.outer(np.cos(u), np.cosh(v))
y = ry * np.outer(np.sin(u), np.cosh(v))
z = rz * np.outer(np.ones_like(u), np.sinh(v))

my_cmap = plt.get_cmap('rainbow')

surf = ax.plot_surface(x,y,z,rstride = 10,cstride = 10,alpha = 0.6,cmap = my_cmap)

cset = ax.contourf(x,y,z,zdir ='z',offset = -2,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='x',offset =-4,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='y',offset = 2,cmap = my_cmap)

ax.set_xlabel('X-axis'); ax.set_xlim(-4, 4)
ax.set_ylabel('Y-axis'); ax.set_ylim(-2, 2)
ax.set_zlabel('Z-axis'); ax.set_zlim(-2, 2)

ax.set_title('График однополостного гиперболоида $2x^2+2y^2-2z^2=1$ с проекциями')

ax.view_init(15, -40)

plt.show()
```

Упражнение 8.4. Построить параболический цилиндр x^2-2x=2z. <br>Изобразить проекцию данной поверхности на плоскость Oxz (используйте функцию ax.view_init(0, ±90)).


```python
fig = plt.figure(figsize =(7, 7))
ax = plt.axes(projection ='3d')
u = np.linspace(-3, 3, 70)
h = np.linspace(0, 2, 70)

# (x - 1)^2 / 2 - 0.5 = z
x = np.outer(u, np.ones(len(h)))
y = np.outer(np.ones(len(h)), h)
z = np.outer ((u-1)**2 / 2 - 0.5, np.ones(len(h)))

ax.plot_surface(x, y, z, alpha=0.8, cmap=plt.get_cmap('gnuplot'))

cset = ax.contourf(x,y,z,zdir ='z',offset = -2,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='x',offset =-4,cmap = my_cmap)
cset = ax.contourf(x,y,z,zdir ='y',offset = 2,cmap = my_cmap)

ax.set_xlabel('X-axis'); ax.set_xlim(-4, 4)
ax.set_ylabel('Y-axis'); ax.set_ylim(-2, 2)
ax.set_zlabel('Z-axis'); ax.set_zlim(-2, 2)

ax.set_title('График параболического цилиндра $x^2-2x=2z$ с проекциями')

# ax.view_init(0, 90)

plt.show()
```

Упражнение 8.5. Найти точки пересечения поверхности x^2-2^x=2z и прямой, проходящей через точки (8, –1, –3) и (–4,3,3). <br>Сделать рисунок. Нанести на него найденные точки.

```python
L = Line(Point(8, -1, -3), Point(-4, 3, 3))
print(L.equation()[0], ' = 0')

# вводим параметризацию
L = L.arbitrary_point()
print(L)

x, y, z, t = symbols('x y z t')

P = x**2 - 2*x - 2*z

eq = P.subs([(x, L[0]), (z, L[2])])

res = solve(eq)
print('t = ', res)

t = res[0]
m1 = Point((-12*t + 8), (4*t - 1), (6*t - 3))
t = res[1]
m2 = Point((-12*t + 8), (4*t - 1), (6*t - 3))

print(f'm1: {m1}\nm2: {m2}')

# Построение цилиндра
fig = plt.figure(figsize =(5, 5))
ax = plt.axes(projection ='3d')
u = np.linspace(-3, 3, 70)
h = np.linspace(-3, 4, 70)

x = np.outer(u, np.ones(len(h)))
y = np.outer(np.ones(len(h)), h)
z = np.outer ((u-1)**2 / 2 - 0.5, np.ones(len(h)))

ax.plot_wireframe(x, y, z, cmap=plt.get_cmap('gnuplot'),
                  color='grey', alpha=0.6)

ax.plot([8, -4], [-1, 3], [-3, 3])
ax.plot(*m1, 'bo')
ax.plot(*m2, 'bo')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
# Построение прямой по двум точкам (при t=-1/2 и t=1/2)
ax.set_xlabel('X-axis'); ax.set_xlim(-4, 4)
ax.set_ylabel('Y-axis'); ax.set_ylim(-2, 2)
ax.set_zlabel('Z-axis'); ax.set_zlim(-2, 2)

ax.set_title('График параболического цилиндра $x^2-2x=2z$')

ax.view_init(25, 80)

plt.show()

```
