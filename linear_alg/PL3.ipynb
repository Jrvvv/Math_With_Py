Упражнение 3.2. Доказать, что векторы  образуют базис. Изобразить эти векторы. 

```python
a = np.array([1, -2, 0])
b = np.array([0, 1, 1])
c = np.array([1, 2, 2])

# Докажем, что они не компларанры

# Создаем матрицу перехода из векторов a, b и c
# и вычисляем её определитель
matrix = np.vstack((a, b, c))

det = np.linalg.det(matrix)

if det != 0:
    print("Векторы образуют базис")
else:
    print("Векторы не образуют базис")

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlim([-1, 2])
ax.set_ylim([-3, 3])
ax.set_zlim([-1, 3])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

start = np.array([0, 0, 0])

ax.quiver(start[0], start[1], start[2], a[0], a[1], a[2], label='a', color='blue')
ax.quiver(start[0], start[1], start[2], b[0], b[1], b[2], label='b', color='red')
ax.quiver(start[0], start[1], start[2], c[0], c[1], c[2], label='c', color='green')
# ax.view_init(2, 2) # for different view

plt.legend()

plt.show()
```

Упражнение 3.3. Проверить являются ли векторы  линейно зависимыми и, если возможно, разложить вектор  по этим векторам (при решении системы использовать формулы Крамера или обратную матрицу). Векторы задать в символьном виде.<br>


```python
# задаем символьные переменные
a, b, c = symbols('a b c')

# исходные векторы
p = a - b + c
q = b + a
r = b - c
s = a + b + c

# формируем матрицу из коэффициентов
M = Matrix([[p.coeff(i), q.coeff(i), r.coeff(i)] for i in [a, b, c]])
print(M)

# проверяем линейную независимость векторов p, q и r
if M.det() != 0:
    print('Векторы p, q, и r линейно независимы.')

    M_inv = M.inv()
    
    # находим решение системы через обратную матрицу (@ - умножение векторов)
    #  A * X = B  ---->  X = A**-1 * B
    abc = M_inv @ Matrix([s.coeff(i) for i in [a, b, c]])
    
    # выводим разложение вектора s по векторам p, q, и r
    print(f"Вектор s можно выразить как: {abc[0]} * p + {abc[1]} * q + {abc[2]} * r")
else:
    print('Векторы p, q, и r are линейно зависимы, невозможно выразить s.')
```

```python
# Пример 2. Найти угол между векторами (0,5,–1) и (–4,9,3).

a = np.array([0, 5, -1])
b = np.array([-4, 9, 3])
print ('Вектор a:', a)
print ('Вектор b:', b)
cos_angle = a @ b / np.linalg.norm(a) / np.linalg.norm(b)
print ('Косинус угла между a и b:', cos_angle)
alfa= np.arccos(cos_angle) # арккосинус угла
alfa=math.degrees(alfa)
print ('Угол в градусах:', alfa)
```

Упражнение 3.4. Даны три точки A(–2,0), B(3,4), C(4,–1). Найти величины углов треугольника АВС.<br>Проверить свойство суммы углов треугольника. Сделать рисунок.

```python
# Задаем координаты точек A, B, C
A = np.array([-2, 0])
B = np.array([3, 4])
C = np.array([4, -1])

# Вычисляем векторы сторон треугольника
AB = B - A
BC = C - B
CA = A - C

# Вычисляем длины сторон треугольника
AB_norm = np.linalg.norm(AB)
BC_norm = np.linalg.norm(BC)
CA_norm = np.linalg.norm(CA)

# Находим значения косинусов углов
# cos(a) = vec_a * vec_b / (norm_a * norm_b)
cos_BAC = AB @ CA / AB_norm / CA_norm
cos_ABC = AB @ BC / AB_norm / BC_norm
cos_ACB = CA @ BC / CA_norm / BC_norm

# Находим значения углов
# Переводим значения в градусы
angle_A_deg = np.degrees(np.arccos(cos_BAC))
angle_B_deg = np.degrees(np.arccos(cos_ABC))
angle_C_deg = np.degrees(np.arccos(cos_ACB))

# Выводим значения углов треугольника
print("Угол А: ", angle_A_deg, "градусов")
print("Угол B: ", angle_B_deg, "градусов")
print("Угол C: ", angle_C_deg, "градусов")

# Проверяем свойство суммы углов треугольника
if angle_A_deg + angle_B_deg + angle_C_deg == 180:
    print("Сумма углов треугольника равна 180 градусов")
else:
    print("Сумма углов треугольника не равна 180 градусов")

plt.plot([A[0], B[0]], [A[1], B[1]])
plt.plot([B[0], C[0]], [B[1], C[1]])
plt.plot([C[0], A[0]], [C[1], A[1]])

plt.scatter([A[0], B[0], C[0]], [A[1], B[1], C[1]])

plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k') 

plt.xlabel('x')
plt.ylabel('y')

plt.grid()

plt.show()
```

Упражнение 3.6. Вычислить площадь треугольника АВС с вершинами A(1,3,–1), B(2,–1,4), C(5,0,3).<br>Как соотносятся площадь треугольника и модуль векторного произведения?

```python
# Задаем координаты точек треугольника
A = np.array([1,  3, -1])
B = np.array([2, -1,  4])
C = np.array([5,  0,  3])

# Вычисляем векторы сторон треугольника
AB = B - A
BC = C - B
CA = A - C

# Вычисляем длины сторон треугольника
AB_norm = np.linalg.norm(AB)
BC_norm = np.linalg.norm(BC)
CA_norm = np.linalg.norm(CA)

# Вычисляем площадь треугольника
p = (AB_norm + BC_norm + CA_norm) / 2
S = np.sqrt(p * (p - AB_norm) * (p - BC_norm) * (p - CA_norm))
print('Площадь: ', S)

# Вычисляем модуль векторного произведения
C_mod = np.linalg.norm(np.cross(AB, BC))
print('Модуль векторного произведения: ', C_mod)

print('S/|C| = ', (S / C_mod).round(1))
print('Поскольку модуль векторного проивзедения'
      ' равен площади параллелограмма')
```

Упражнение С1. Используя определение смешанного произведения, вычислить объем пирамиды ABCD, если<br><br>Проверить с помощью определителя.

```python
# Задаем координаты точек A, B, C, D
A = np.array([0, 0, 0])
B = np.array([3, 4, 0])
C = np.array([-3, 0, 1])
D = np.array([0, 1, 5])

# Вычисляем векторы сторон пирамиды
AB = B - A
AC = C - A
AD = D - A

# Вычисляем смешанное произведение векторов AB, AC, AD
# v_a * v_b * v_c = (v_a x v_b) * v_c = -(v_b x v_a) * vec_c
V = AB @ np.cross(AC, AD) # объем параллепипеда

# Вычисляем объем пирамиды по формуле V = 1/6 * |V|
V = abs(V) / 6

print("Объем пирамиды ABCD: ", V)

# Проверяем результат с помощью определителя матрицы из координат векторов AB, AC, AD
# det(M) = v_a * v_b * v_c;   M = [a, b. c]
M = np.array([AB, AC, AD])
det_M = (np.linalg.det(M)).round(6)

# Выводим определитель матрицы
print("Определитель матрицы: ", det_M)

# Проверяем результаты вычислений
if V == abs(det_M) / 6:
    print("Вычисления верны")
else:
    print("Вычисления неверны")
```
