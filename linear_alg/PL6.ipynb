Произвести сдвиг начала координат в точку (3,2) и поворот графика функции y = 2x - 4<br> относительно начала координат на 45* против часовой стрелки (положительное направление).<br> Найти уравнение полученной прямой и координаты точки М(1,–2).<br> Изобразить прямые и точки на графике.

```python
a = 3
b = 2

# создаём символьные переменные
# искомое уравнение – это уравнение прямой в новой системе координат
x, y, alpha = sp.symbols('x y alpha')

# Новые координаты через старые
x1 = (x - a) * sp.cos(-alpha) - (y - b) * sp.sin(-alpha)
y1 = (y - b) * sp.cos(-alpha) + (x - a) * sp.sin(-alpha)

f = y1 - 2 * x1 + 4

# решаем систему уравнений, выражая у
y1 = sp.solve(f, 'y')[0]
print('Решение системы: у = ', y)

yy = y1.subs(alpha, np.pi / 4)
print('Уравнение повернутой прямой: у = ', yy)

# строим прямые
x_val = np.linspace(-2, 9, 100)
y_val = [yy.subs(x, m) for m in x_val] # выражаем y

plt.plot(x_val, y_val, label='y = 16.66 - 3x') # в новой системе
plt.plot(x_val, 2 * x_val - 4, label='y = 2x - 4')  # в старой системе

# обрабатываем точку
alpha = np.pi / 4
x0 = 1
y0 = -2

# Новые координаты через старые
x2 = x0 * np.cos(alpha) - y0 * np.sin(alpha) + a
y2 = y0 * np.cos(alpha) + x0 * np.sin(alpha) + b
print('Координаты точки в новой системе', x2, y2)

# строим точки
plt.plot(x0, y0, 'ok', markersize=5, label='M0(1, -2)') # в старой системе
plt.plot(x2, y2, 'ob', markersize=5, label=f'M1({x2.round(3)}, {y2.round(3)})') # в новой системе

# координатные оси
plt.arrow(-2, 0, 6, 0, width=0.03, color='black')
plt.arrow(0, -2, 0, 5, width=0.03, color='black')

# координаты нового базиса
x3 = np.cos(-alpha); y3 = -np.sin(-alpha)
x4 = np.sin(-alpha); y4 = np.cos(-alpha)

# координаты нового базиса
plt.arrow(a, b, x3, y3, width=0.03, color='blue')
plt.arrow(a, b, x4, y4, width=0.03, color='blue')

plt.xlabel('x')
plt.ylabel('y')

plt.grid()

plt.axis('equal')

plt.axis([-1, 4, -4, 4])

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',
           borderaxespad=0.0, shadow=True)
```
