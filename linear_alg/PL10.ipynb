Упражнение 10.1. Найти матрицу перехода T, которая приводит матрицу А<br><br>
к диагональному виду. Сделать проверку: A = T * D * T-1.


```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

A = Matrix([[1, -3, 1],
            [3, -3, -1],
            [3, -5, 1]])

T, D = A.diagonalize()

print('Матрица перехода:')
pprint(T)

print('Диагональная матрица:')
pprint(D)

print('Проверка:')
pprint(T * D * T.inv())
```

Упражнение 10.3. Определить тип кривой 2x^2 - 4xy + 5y^2 + 8x - 2y + 9 = 0.<br>Найти каноническое уравнение и записать формулы перехода к новым переменным. <br>Построить кривую в новой системе координат.


```python
A = Matrix([[2, -2], [-2, 5]])

T, D = A.diagonalize()

print('Диагноализированная матрица:')
pprint(D)

x1, y1 = symbols('x1 y1')

X1 = Matrix([[x1, y1]])

X = T/sqrt(5) * X1.T

print('x = ', simplify(X[0, 0]))
print('y = ', simplify(X[1, 0]))

F = 2*X[0]**2 - 4*X[0]*X[1] + 5*X[1]**2 + 8*X[0] - 2*X[1] + 9
F = simplify(F)

print()
pprint(F)

var('x y')
plot_implicit(Eq(2*x**2 - 4*x*y + 5*y**2 + 8*x - 2*y, -9), title='В старой СК')
plot_implicit(Eq(x**2 + 6*y**2 + 14*sqrt(5)*x/5 - 12*sqrt(5)*y/5, -9), title='В новой СК')
```

Упражнение 10.4. Определить тип кривой x^2+4xy+4y^2-6x-2y+1=0.<br>Найти каноническое уравнение и записать формулы перехода к новым переменным. <br>Построить кривую в старой и новой системах координат.


```python
A = Matrix([[1, 2], [2, 4]])

T, D = A.diagonalize()

print('Диагноализированная матрица:')
pprint(D)

x1, y1 = symbols('x1 y1')

X1 = Matrix([[x1, y1]])

X = T/sqrt(5) * X1.T

print('x = ', simplify(X[0, 0]))
print('y = ', simplify(X[1, 0]))

F = X[0]**2 + 4*X[0]*X[1] + 4*X[1]**2 - 6*X[0] - 2*X[1] + 1
F = simplify(F)

print()
pprint(F)

var('x y')
plot_implicit(Eq(x**2 + 4*x*y + 4*y**2 - 6*x - 2*y, -1), title='В старой СК')
plot_implicit(Eq(2*sqrt(5)*x + 5*y**2 - 2*sqrt(5)*y, -1), title='В новой СК')
```

Упражнение 10.5. Определить тип поверхности x^2 - 2y^2 + z^2 + 4xy - 8xz - 4yz + 6 = 0 и построить ее в новой системе координат.

```python
x1, y1, z1 = symbols('x1 y1 z1')

A = Matrix([[1, 2, -4],
            [2, -2, -2],
            [-4, -2, 1]])

T1, D1 = A.diagonalize()

XX = Matrix([[x1,y1,z1]])

X = T1 * XX.T # матрица Т1 - матрица перехода из упр.10.2

pprint(D1)
print('x=',simplify(X[0,0]))
print('y=',simplify(X[1,0]))
print('z=',simplify(X[2,0]))

H = X[0]**2 - 2*X[1]**2 + X[2]**2 + 4*X[0]*X[1] - 8*X[0]*X[2] - 4*X[1]*X[2] + 6
H = simplify(H)
print()
pprint(H/6)
print('Канонический вид: -x^2 + 0.5y^2 + 0.5z^2 = 1')

inv_3 = Matrix([[-15, 3, 0, 0],
                [3, -6, 0, 0],
                [0, 0, 54, 0],
                [0, 0, 0, 6]]).det()

print('I3 = ', inv_3) # собюственные значения разных знаков
                      # 3 инвариант > 0

print('Фигура -- однополостный гиперболоид')

# Построение эллипсоида в сферических координатах
fig = plt.figure(figsize=(7, 7)) # создаём холст

ax = fig.add_subplot(111, projection='3d')

#Коэффициенты уравнения
# -x^2 + 0.5y^2 + 0.5z^2 = 1
coefs = (1, 2, 2)

rx, ry, rz = np.sqrt(coefs)

u = np.linspace(0, 2*np.pi, 200) # Угол на плоскости
v = np.linspace(-np.pi/2, np.pi/2, 200) # Угол в пространстве

#Уравнение эллипсоида
x = rx * np.outer(np.cos(u), np.cosh(v))
y = ry * np.outer(np.sin(u), np.cosh(v))
z = rz * np.outer(np.ones_like(u), np.sinh(v))

my_cmap = plt.get_cmap('rainbow')

surf = ax.plot_surface(x,y,z,rstride = 10,cstride = 10,alpha = 0.6,cmap = my_cmap)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Однополостный гиперболоид')

plt.show()
```
