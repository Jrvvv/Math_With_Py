```python
x = 1234.56789

# Два десятичных знака точности после запятой
format(x, '0.2f')
#'1234.57'
```

```python
# Выравнивание по правому краю в 10 символов, один знак точности 
format(x, '>10.1f')
#'    1234.6'
```

```python
# Выравнивание по левому краю 
format(x, '<10.1f')    
#'1234.6    '
```

```python
# Выравнивание по центру 
format(x, '^10.1f')
#'  1234.6  '
```

```python
# Включение разделителя разрядов 
format(x,',')
#'1,234.56789'
```

```python
format(x, '0,.1f')
#'1,234.6'
```

```python
format(x, 'e')
#'1.234568e+03'
```

```python
format(x, '0.2E')
#'1.23E+03'
```

Упражнение 4. Вычислите значения следующих переменных:<br>
<br>и отобразите результата в виде столбца чисел с тремя знаками после запятой и десятичными точками друг над другом.

```python
x = 1/3 - 1/7
y = 2**4 + 2**(-4)
z = (3 + 0.2) / (1 - 0.35)

print('x =', format(x, '>6.3f'))
print('y =', format(y, '>6.3f'))
print('z =', format(z, '>6.3f'))
```
