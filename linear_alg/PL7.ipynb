Упражнение 7.1. Построить эллипс с центром в точке (3,1), горизонтальной полуосью, равной 3, и эксцентриситетом, равным 4/5. <br>Нанести на рисунок центр эллипса и его фокусы. Найти все параметры эллипса и его параметрическое и каноническое уравнения.


```python
el = sp.Ellipse(sp.Point(3, 1), hradius=3, eccentricity=sp.Rational(4, 5))

a = el.hradius
b = el.vradius
E = el.eccentricity
F = el.foci

print('a = ', a,'\nb = ', b, '\nE = ', E, '\nF = ', F)
print('Каноническое уравнение: ', el.equation())
print(f'Параметрическое уравнение:\nx = {a}cos(t) + 3\ny = {b}*sin(t) + 1')

t = np.arange(0, np.pi * 2, 0.01)
x = a * np.cos(t) + 3
y = b * np.sin(t) + 1

plt.plot(x, y, lw=2)

plt.plot(3, 1, 'o')

plt.plot(F[0][0], F[0][1], 'ro')
plt.plot(F[1][0], F[1][1], 'ro')

plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k') 

plt.ylabel('$y$')
plt.xlabel('$x$')

plt.axis('equal')

plt.grid()
plt.show()
```

Упражнение 7.2. Построить параболу с фокусом в точке (0,0) и директрисой y = 8:

```python
x, y = sp.symbols('x y')

pl = sp.Parabola(sp.Point(0, 0), sp.Line(sp.Point(0, 8), sp.Point(1, 8)))
F = pl.focus

pl_eq = pl.equation(x, y)
print('Каноническое уравнение: ', pl_eq, '= 0')

f = sp.solve(pl_eq, y)[0]

x_val = np.arange(-10, 10, 0.01)

y_val = [float(f.subs(x, m).evalf()) for m in x_val] # подставляем x

plt.plot(x_val, y_val)

plt.axhline (y=8, color='r')

plt.plot(F[0], F[1], 'ro')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.ylabel('$y$')
plt.xlabel('$x$')

plt.axis('equal')

plt.grid()
plt.show()
```

Упражнение 7.3. Построить кривую y^2/2-(x-2)^2/8=1, ее асимптоты и фокусы.

```python
x, y = sp.symbols('x y')

exp = y**2/2 - (x-2)**2/8 - 1

f1 = sp.solve(exp, y)[0]
f2 = sp.solve(exp, y)[1]

print(f'y1 = {f1}\ny2 = {f2}')

a = np.sqrt(8)
b = np.sqrt(2)
c = np.sqrt(2 + 8)

x_val = np.linspace(-30, 30, 100)

y1_val = [float(f1.subs(x, m).evalf()) for m in x_val] # подставляем x
y2_val = [float(f2.subs(x, m).evalf()) for m in x_val] # подставляем x

y1_asympt = (x_val - 2) * b / a 
y2_asympt = (x_val - 2) * -b / a

plt.plot(x_val, y1_val, color='blue')
plt.plot(x_val, y2_val, color='blue')

plt.plot(x_val, y1_asympt, color='red')
plt.plot(x_val, y2_asympt, color='red')

plt.plot(2, c, 'go')
plt.plot(2, -c, 'go')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.ylabel('$y$')
plt.xlabel('$x$')

plt.axis('equal')

plt.grid()
plt.show()
```

Упражнение 7.4. Найти параметрический вид кривой y^2/2-(x-2)^2/8=1. Построить эту кривую (возьмите t[-3;3]). Показать, как меняются координаты точки в зависимости от параметра t. 


```python
a = np.sqrt(8)
b = np.sqrt(2)
c = np.sqrt(2 + 8)

x, y, t = sp.symbols('x y t')

x_eq = a.round(2) * sp.cosh(t) - 2
y_eq = b.round(2) * sp.sinh(t)

print(f'x = {x_eq}\ny = {y_eq}')

t_val = np.linspace(-3, 3, 75)

x1_val = a * np.cosh(t_val) + 2
x2_val = -a * np.cosh(t_val) + 2

y_val = np.sinh(t_val) * b

plt.plot(y_val, x1_val, 'or')
plt.plot(y_val, x2_val, 'og')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.ylabel('$y$')
plt.xlabel('$x$')

plt.axis('equal')

plt.grid()
plt.show()
```

Упражнение 7.5. Найти точки пересечения параболы x^2=y+1 и эллипса x^2/4+y^2/25=1. Сделать рисунок. Нанести на рисунок полученные точки.




```python
x, y = sp.symbols('x y')

a = 2
b = 5

p = sp.Parabola(sp.Point(0,0), sp.Line(sp.Point(1, -2), sp.Point(-1,-2)))
el = sp.Ellipse(sp.Point(0, 0), a, b)

points = p.intersection(el)
print(points)

# точки эллипса
t = np.arange(0, 2*np.pi, 0.05)
x_el = a*np.cos(t)
y_el = b*np.sin(t)

# точки параболы
p_eq = p.equation(x, y)

f = sp.solve(p_eq, y)[0]

x_par = np.arange(-7, 7, 0.01)

y_par = [float(f.subs(x, m).evalf()) for m in x_par] # подставляем x

# Отрисовка
plt.plot(x_el, y_el, lw=2)
plt.plot(x_par, y_par)

plt.plot(points[0][0], points[0][1], 'o')
plt.plot(points[1][0], points[1][1], 'o')

plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k') 

plt.ylabel('$y$')
plt.xlabel('$x$')

plt.axis('equal')

plt.grid()
plt.show()
```

Упражнение 7.6. Определить тип и привести к каноническому виду кривую второго порядка 4x^2+9y^2-8x-36y+4=0. <br>Найти фокус, фокальное расстояние, эксцентриситет. Сделать проверку.

```python
x, y = sp.symbols('x y')

f = (4*x**2 + 9*y**2 - 8*x - 36*y + 4)/36

sp.pprint(sp.sqrt(f))
print()
sp.pprint(sp.solve(f, x**2)[0])
print()
sp.pprint(sp.solve(f, y**2)[0])
print()

el = sp.Ellipse(sp.Point(1, 2), 3, 2)
sp.pprint(sp.expand(el.equation()))

print('Координаты фокусов:', el.foci)
print('Фокусное расстояние:', el.focus_distance)
print('Эксцентриситет:', el.eccentricity)
```
