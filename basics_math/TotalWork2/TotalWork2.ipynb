2. Построить график указанной функции, номер которой совпадает с вашим номером в
списке группы. Найти нули функции, точки экстремума и значения в них, точки перегиба,
значения в них, значения тангенса угла наклона касательной в точке перегиба, найти
односторонние пределы в точках разрыва, уравнения асимптот. Обозначить на графике
экстремумы, построить касательные в окрестностях точек перегиба, асимптоты. Задание
сдается в бумажном виде на одной странице.

```python
import numpy as np
import sympy as sp
from scipy.optimize import minimize_scalar

# Определяем символьные переменные и функцию
x = sp.Symbol('x', real=True)
y = (x**3 + 2*(x**2)) / (x - 2)

# 1) Находим нули функции, точки экстремума и значения в них
roots = [float(r.evalf()) for r in sp.solve(y, x)]
extrema_points = [float(p.evalf()) for p in sp.solve(sp.diff(y), x)]
extrema_values = [float(y.subs({x: p}).evalf()) for p in extrema_points]

# 2) Находим точки перегиба и значения в них
inflection_points = [float(p.evalf()) for p in sp.solve(sp.diff(y, x, 2), x)]
inflection_value = float(y.subs({x: inflection_points[0]}).evalf())

# 3) Находим значения тангенса угла наклона касательной в точке перегиба
slope_tangent_inflection = float(sp.diff(y).subs({x: inflection_points[0]}).evalf())

# 4) Находим односторонние пределы в точках разрыва
limit_left = float(sp.limit(y, x, 2, dir="-").evalf())
limit_right = float(sp.limit(y, x, 2, dir="+").evalf())

# 5) Находим уравнения асимптот
asymptote_vertical = sp.Eq(x, 2)
div = sp.div(y, x**2)
asymptote_oblique = sp.Eq(y, x + 4)

print("Точки перегиба: ", inflection_points)
print("Значение функции в точке перегиба: ", inflection_value)
print("Нули функции: ", roots)
print("Точки экстремума: ", extrema_points)
print("Значения функции в точках экстремума: ", extrema_values)
print("Точки перегиба: ", inflection_points)
print("Значение функции в точке перегиба: ", inflection_value)
print("Тангенс угла наклона касательной в точке перегиба: ", slope_tangent_inflection)
print("Левосторонний предел: ", limit_left)
print("Правосторонний предел: ", limit_right)
print("Уравнение вертикальной асимптоты: ", asymptote_vertical)
print("Уравнение наклонной асимптоты: ", asymptote_oblique)

# Построение графиков
x=np.linspace(-20,20,1000)
plt.figure(figsize=(8, 8))

x_orig = np.linspace(-20,20,1000)
y_orig = (x_orig**3 + 2*(x_orig**2)) / (x_orig - 2)

handle1, = plt.plot(x_orig, y_orig, color='b', linewidth=2.0, label='y = (x^3 + 2*x^2) / (x - 2)')

# построение асимптоты
plt.axvline(x=2,  ymin=-5, ymax=49, color='r',
          linewidth=3.0, linestyle='--', label='x = 2 асимптота') 

# построение касательной в точке перегиба
x_tangent = np.linspace(-20,20,1000)
y_tangent = slope_tangent_inflection * x - (inflection_value - slope_tangent_inflection * inflection_points[0])
handle2, = plt.plot(x_tangent, y_tangent, color='g', label='y = 0,44*x - 0,07\nкасательная в точке перегиба (-0.52, -0.15)')

plt.axis('square')
plt.xlim((0-15,0+15))
plt.ylim((0-5/np.sqrt(2),0+80/np.sqrt(2)))
plt.legend(handles=[handle1, handle2])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',
           borderaxespad=0.0, shadow=True)

plt.grid(True)
plt.axvline(x = 0, color = 'k')
plt.axhline(y = 0, color = 'k')

#Точки экстремума (точки)
plt.plot(extrema_points[0], extrema_values[0] ,'or', markersize=5)
plt.plot(extrema_points[1], extrema_values[1] ,'or', markersize=5)
plt.plot(extrema_points[2], extrema_values[2] ,'or', markersize=5)

#Точки экстремума (значения)
plt.text(extrema_points[0] + 1 , extrema_values[0] - 1, 'max(0; 0)')
plt.text(extrema_points[1] - 4 , extrema_values[1] - 2, 'min(-1,24; -0,36)')
plt.text(extrema_points[2] + 1 , extrema_values[2] - 1.5, 'min(3,24; 44,36)')

plt.axis([-15,20,-4,50])

plt.show()
```
