```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cmath as cmt
```

```python
x = np.linspace(-1, 1,100)
handle1, = plt.plot(x, x**2, label='y=x**2')
plt.title('$x**2;     +-sqrt(x)$')
x = np.linspace(-1, 1,100)
handle2, = plt.plot(x, np.sqrt(x), 'g--', label='y=sqrt(x)')
handle3, = plt.plot(x, -np.sqrt(x),  'g--', label='y=-sqrt(x)')
plt.plot(x, x, 'r-.')
plt.legend(handles=[handle1, handle2, handle3])
plt.axis('square')
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k') 
plt.grid()
plt.show()
```

```python
x = np.linspace(-1, 1,100)
handle1, = plt.plot(x,x**(3), label='y=x**3')
plt.title('$x**3;     cbrt(x)$')
x = np.linspace(-1, 1,100)
handle2, = plt.plot(x, np.cbrt(x), 'g--', label='y=cbrt(x)')
plt.plot(x, x, 'r-.')
plt.legend(handles=[handle1, handle2])
plt.axis('square')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k') 
plt.grid()
plt.show()
```

```python
x=np.linspace(0.15,5,100)
handle1, = plt.plot(x,np.log(x),'b-', label='log(x)')
x=np.linspace(-2,1.5,100)
handle2, = plt.plot(x,np.exp(x),'g--', label='exp(x)')
x=np.linspace(-2,4.5,100)
plt.title('$ln(x);     exp(x)$')
plt.plot(x,x,'r-.')
plt.legend(handles=[handle1, handle2])
plt.axis('square')
plt.grid()
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k') 
```

```python
x = np.linspace(-2, 3.9,100)
handle1, = plt.plot(x, 0.5**x, label='y=0,5**(x)')
plt.title('$0,5**(x);     log0,5(x)$')
x = np.linspace(-4, 4,100)
handle2, = plt.plot(x, (np.log10(x)/np.log10(0.5)), 'g--', label='y=log0,5(x)')
plt.plot(x, x, 'r-.')
plt.legend(handles=[handle1, handle2])
plt.axis('square')
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k') 
plt.grid()
plt.show()
```

```python
x = np.linspace(-2, 2,100)
handle1, = plt.plot(x,np.sin(x), label='y=sin(x)')
plt.title('$sin(x);    arcsin(x) $')
x = np.linspace(-2, 2,1000)
handle2, = plt.plot(x, np.arcsin(x), 'g--', label='y=arcsin(x)')
plt.legend(handles=[handle1, handle2])
plt.plot(x, x, 'r-.')
plt.axis('square')
plt.axis(xmin=-2, xmax=2, ymin=-2, ymax=2)
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k') 
plt.grid()
plt.show()
```

```python
x = np.linspace(-4, 4,100)
handle1, = plt.plot(x,np.cos(x), label='y=cos(x)')
plt.title('$cos(x);    arccos(x) $')
x = np.linspace(-4, 4,10000)
handle2, = plt.plot(x, np.arccos(x), 'g--', label='y=arcsin(x)')
plt.plot(x, x, 'r-.')
plt.legend(handles=[handle1, handle2])
plt.axis('square')
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k') 
plt.grid()
plt.show()
```

```python
x = np.linspace(-1.3, 1.3 ,100)
handle1, = plt.plot(x,np.tan(x), label='y=tg(x)')
plt.title('$tg(x);    arctg(x) $')
x = np.linspace(-4, 4,10000)
handle2, = plt.plot(x, np.arctan(x), 'g--', label='y=arctg(x)')
plt.plot(x, x, 'r-.')
plt.legend(handles=[handle1, handle2])
plt.axis('square')
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k') 
plt.grid()
plt.show()
```

```python
x=np.linspace(-4,4,10000)
handle1, = plt.plot(x, 1/np.tan(x),'b', label='y=ctg(x)')
plt.plot(x,x,'r-.')
x=np.linspace(0,3,10000)
handle2, = plt.plot(1/np.tan(x), x,'g--', label='y=arcctg(x)')
plt.legend(handles=[handle1, handle2])
plt.title('$ctg(x);    arctg(x) $')
plt.axis(xmin=-5, xmax=5, ymin=-3, ymax=4)
plt.axhline(y=0, color='k') 
plt.axvline(x=0, color='k') 
plt.grid()
plt.show()

```
