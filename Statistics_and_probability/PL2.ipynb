Для каждого из следующих трёх видов распределений: <br>
- нормального N(m; s) ( s > 0 ),<br>
- равномерного R(a; b) ( a < b ),<br>
- экспоненциального E(l) ( l > 0 )<br><br>
написать программу вывода в графическом окне по два графика функций плотности вероятности, причём<br>
один из этих графиков получить по собственной программе, написанной для расчёта значений функции<br>
плотности вероятности по формулам раздела, другой – с использованием функций.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, expon

# параметры распределения
m = 20
sigma = np.sqrt(20)

# собственная программа для расчета функции плотности вероятности
def my_norm(x, m, sigma):
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-m)**2/(2*sigma**2))

fig = plt.figure(figsize=(8, 8))

# график функции плотности вероятности, полученный с помощью собственной программы
x = np.linspace(m-4*sigma, m+4*sigma, 1000)
y_my = my_norm(x, m, sigma)
plt.plot(x, y_my, label='my', linewidth=4)

# график функции плотности вероятности, полученный с помощью функции из библиотеки scipy
y_scipy = norm.pdf(x, m, sigma)
plt.plot(x, y_scipy, label='scipy')

plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid('square')

plt.title('Normal distribution')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k') 

plt.legend()
plt.show()
```

```python
# параметры распределения
a = 20 # минимальный элемент распределения
b = a * 2 # максимальный элемент распределения

# собственная программа для расчета функции плотности вероятности
def my_uniform(x, a, b):
    if x < a or x > b:
        return 0
    else:
        return 1/(b-a)

# график функции плотности вероятности, полученный с помощью собственной программы
x = np.linspace(0, b+5, 1000)
y_my = [my_uniform(i, a, b) for i in x]
plt.plot(x, y_my, label='my', linewidth=4)

# график функции плотности вероятности, полученный с помощью функции из библиотеки scipy
y_scipy = uniform.pdf(x, a, b-a)
plt.plot(x, y_scipy, label='scipy')

plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid('square')

plt.title('Uniform distribution')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k') 

plt.legend()
plt.show()
```

```python
# параметры распределения
lambd = 4

# собственная программа для расчета функции плотности вероятности
def my_expon(x, lambd):
    if x < 0:
        return 0
    else:
        return lambd*np.exp(-lambd*x)

# график функции плотности вероятности, полученный с помощью собственной программы
x = np.linspace(0, 10, 1000)
y_my = [my_expon(i, lambd) for i in x]
plt.plot(x, y_my, label='my', linewidth=4)

# график функции плотности вероятности, полученный с помощью функции из библиотеки scipy
y_scipy = expon.pdf(x, scale=1/lambd)
plt.plot(x, y_scipy, label='scipy')

plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid('square')

plt.title('Exp distribution')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k') 

plt.legend()
plt.show()
```

Для каждого из следующих трёх видов распределений: <br><br>

- хи^2 с k степенями свободы,<br>
- Стьюдента с k степенями свободы,<br>
- Фишера с k1, k2 степенями свободы<br><br>

вывести в графическом окне по два графика функций плотности вероятности, причём один из графиков<br>
получить по собственной программе, написанной для расчёта значений функции плотности вероятности,<br>
другой – с использованием функций.

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.special import gamma
from scipy.stats import chi2
from scipy.stats import t
from scipy.stats import f

# параметры распределения
k = 20

# собственная программа для расчета функции плотности вероятности
def my_khi2(x, k):
    if x < 0:
        return 0
    else:
        return 1/(2**(k/2)*gamma(k/2))*x**(k/2-1)*np.exp(-x/2)

# график функции плотности вероятности, полученный с помощью собственной программы
x = np.linspace(0, 35, 1000)
y_my = [my_khi2(i, k) for i in x]
plt.plot(x, y_my, label='my', linewidth=4)

# график функции плотности вероятности, полученный с помощью функции из библиотеки scipy
y_scipy = chi2.pdf(x, k)
plt.plot(x, y_scipy, label='scipy')

plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid('square')

plt.title('Khi^2 distribution')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k') 

plt.legend()
plt.show()
```

```python
# параметры распределения
k = 20

# собственная программа для расчета функции плотности вероятности
def my_t(x, k):
    return gamma((k+1)/2)/(np.sqrt(k*np.pi)*gamma(k/2))*(1+x**2/k)**(-(k+1)/2)

# график функции плотности вероятности, полученный с помощью собственной программы
x = np.linspace(-5, 5, 1000)
y_my = [my_t(i, k) for i in x]
plt.plot(x, y_my, label='myf', linewidth=4)

# график функции плотности вероятности, полученный с помощью функции из библиотеки scipy
y_scipy = t.pdf(x, k)
plt.plot(x, y_scipy, label='scipy')

plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid('square')

plt.title('Student\'s distribution')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k') 

plt.legend()
plt.show()
```

```python
k1 = 4
k2 = 9

# собственная программа для расчета функции плотности вероятности
def my_f(x, k1, k2):
    if x < 0:
        return 0
    else:
        return gamma((k1+k2)/2)/(gamma(k1/2)*gamma(k2/2))*(k1/k2)**(k1/2)*(x**(k1/2-1))/((1+k1*x/k2)**((k1+k2)/2))

# график функции плотности вероятности, полученный с помощью собственной программы
x = np.linspace(0, 10, 1000)
y_my = [my_f(i, k1, k2) for i in x]
plt.plot(x, y_my, label='my', linewidth=4)

# график функции плотности вероятности, полученный с помощью функции из библиотеки scipy
y_scipy = f.pdf(x, k1, k2)
plt.plot(x, y_scipy, label='scipy')

plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid('square')

plt.title('Fisher\'s distribution')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k') 

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Нормальное распределение
m = 20
sigma = np.sqrt(20)
norm_arr = np.random.normal(m, sigma, size=(300,))
plt.hist(norm_arr, bins=8)
plt.xlabel('x')
plt.ylabel('f')

plt.grid('square')

plt.title('Normal distribution')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k') 

plt.show()

# Равномерное распределение
a = 20
b = 2*a
unif_arr = np.random.uniform(a, b, size=(300,))
plt.hist(unif_arr, bins=8)
plt.xlabel('x')
plt.ylabel('f')

plt.grid('square')

plt.title('Uniform distribution')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k') 

plt.show()

# Экспоненциальное распределение
lam = 4

exp_arr = np.random.exponential(lam, size=(300,))

plt.hist(exp_arr, bins=8)
plt.xlabel('x')
plt.ylabel('f')

plt.grid('square')

plt.title('Exp distribution')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k') 

plt.show()
```
